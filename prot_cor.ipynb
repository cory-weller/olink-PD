# Formal analysis

```python
import pandas as pd
import numpy as np
import os as os
from scipy.stats import pearsonr
```

```python
rm -r data/*
```

```python
Ds=['D01', 'D02']
modes=['cardiometabolic', 'inflammation', 'neurology', 'oncology']
origs=['CSF', 'PLA']
visits=['BLM0T1', 'SVM12T1', 'SVM24T1', 'SVM36T1']
datasets=['PP-', 'PD-']
```

```python
folder_path='/data/CARD/projects/proteomicsXprogression_PD/data-allreleases/amp-pd-v3-proteomics'
for Di in Ds:
    for mode in modes:
        for orig in origs:
            file_path=f"{folder_path}/proteomics-{orig}-PPEA-{Di}/olink-explore/protein-expression/{orig}-PPEA-{Di}_{mode}.csv"
            d=pd.read_csv(file_path)
            for visit in visits:
                for dataset in datasets:
                    t=d[(d.participant_id.str.contains(dataset))&(d.sample_id.str.contains(visit))].copy()
                    t['orig']=orig
                    t2=t[['participant_id', 'orig', 'UniProt', 'NPX']]
                    if t2.shape[0]!=0:
                        t2.to_csv(f'data/{Di}_{mode}_{orig}_{dataset}{visit}.csv', index=False)
                
```

```python
# Correlation calculation between CSF and PLA
dataset='PP-'
visit='BLM0T1'

for Di in Ds:
    for mode in modes:
        f_csf=f'data/{Di}_{mode}_CSF_{dataset}{visit}.csv'
        f_pla=f'data/{Di}_{mode}_PLA_{dataset}{visit}.csv'

        if os.path.exists(f_csf)&os.path.exists(f_pla):
            d_csf=pd.read_csv(f_csf)
            d_pla=pd.read_csv(f_pla)
            common_prot=np.intersect1d(d_csf.UniProt, d_pla.UniProt)


            t1=pd.concat([d_csf, d_pla], ignore_index=True, axis=0)
            t2=t1[t1.UniProt.isin(common_prot)]
            t3=t2.pivot(index='participant_id', columns=['orig', 'UniProt'], values='NPX')

            res=pd.DataFrame()
            
            # Calculate correlation per UniProt
            for UniProt in common_prot:
                t4=t3.iloc[:, t3.columns.get_level_values(1)==UniProt].stack().dropna()
                n=t4.shape[0]
                if n>10:
                    corr, p = pearsonr(t4['CSF'], t4['PLA'])
                    res=pd.concat([res, pd.DataFrame(data={'mode':[mode], 'UniProt':[UniProt], 'n':n, 'corr':[corr], 'p':p})],ignore_index=True)
            res.to_csv(f'data/{Di}_{mode}_COR_{dataset}{visit}.csv', index=False)

        else:
            print('not comparable')
```

```python

# Summarize
for Di in Ds:
    res=pd.DataFrame()
    for mode in modes:
        d=pd.read_csv(f'data/{Di}_{mode}_COR_{dataset}{visit}.csv')
        res=pd.concat([res, d], ignore_index=True)
    res.to_csv(f'data/{Di}_COR_{dataset}{visit}.csv', index=False)
```
